- Preprocess data (Lowercasing and remove stop words)

```python

import json
from string import punctuation
fileName='content.json'
with open(fileName,'r',encoding='utf-8') as fr:
  data=json.load(fr)
def preprocess_text(text):
  text=text.lower()

  stop_word=['thực tập sinh','tts','intern','thực tập']
  for char in text:
    if char in punctuation:
      text=text.replace(char,' ')
  for word in stop_word:
    text=text.replace(word,'')
  return text
preprocessed_data=[preprocess_text(text) for text in data]

with open('processed_data.json','w',encoding='utf-8') as fw:
  json.dump(preprocessed_data,fw,ensure_ascii=False)
print(preprocessed_data)
```

- Dictionaring words


```python
word_dic={}

for text in preprocessed_data:
  for word in text.split():
    if word not in word_dic:
      word_dic[word]=1
    else:
      word_dic[word]+=1
print(word_dic)
```

- Vissualize data


```python
import matplotlib.pyplot as plt

keys = list(word_dic.keys())
values = list(word_dic.values())
plt.figure(figsize=(50, 6))


plt.bar(keys, values,width=0.5)
plt.xlabel('Từ khoá')
plt.ylabel('Số lần xuất hiện')
plt.title('Biểu đồ từ điển')
plt.show()
```
